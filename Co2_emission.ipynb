```python
#remove outliers using z-score

from scipy import stats

z_scores = stats.zscore(data['co2_emissions'])

data = data[abs(z_scores) < 3]
```

```python
from sklearn.preprocessing import OneHotEncoder
x = pd.get_dummies(data.drop(["co2_emissions","model"],axis=1))
y = data['co2_emissions']
```

```python
print(x.shape)
print(y.shape)
```

# Model Building

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

Linear Regression

```python
from sklearn.linear_model import LinearRegression


LR = LinearRegression()
LR.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error

LR_pred = LR.predict(X_test)


# calculate MSE
LR_mse = mean_squared_error(y_test, LR_pred)
print("Mean Squared Error (MSE):", LR_mse)

# calculate RMSE
LR_rmse = mean_squared_error(y_test, LR_pred, squared=False)
print("Root Mean Squared Error (RMSE):", LR_rmse)

#calculate R^2
LR_rsquared = LR.score(X_test, y_test)
print("R^2 Score:", LR_rsquared)
```

KNN

```python
from sklearn.neighbors import KNeighborsRegressor

knn = KNeighborsRegressor()
knn.fit(X_train,y_train)
```

```python
knn_pred = knn.predict(X_test)


# calculate MSE
knn_mse = mean_squared_error(y_test, knn_pred)
print("Mean Squared Error (MSE):", knn_mse)

# calculate RMSE
knn_rmse = mean_squared_error(y_test, knn_pred, squared=False)
print("Root Mean Squared Error (RMSE):", knn_rmse)

#calculate R^2
knn_rsquared = knn.score(X_test, y_test)
print("R^2 Score:", knn_rsquared)
```

Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

RF = RandomForestRegressor()
RF.fit(X_train,y_train)
```

```python
RF_pred = RF.predict(X_test)


# calculate MSE
RF_mse = mean_squared_error(y_test, RF_pred)
print("Mean Squared Error (MSE):", RF_mse)

# calculate RMSE
RF_rmse = mean_squared_error(y_test, RF_pred, squared=False)
print("Root Mean Squared Error (RMSE):", RF_rmse)

#calculate R^2
RF_rsquared = RF.score(X_test, y_test)
print("R^2 Score:", RF_rsquared)
```

# Model Validation

```python
#Evaluate using train test split method

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor


X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=42)


model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
prediction = model.predict(X_test)

mse = mean_squared_error(y_test, prediction)
print("Mean Squared Error:", mse)

rsquared = model.score(X_test, y_test)
print("R^2 Score:", rsquared)
```

```python
#evaluating using cross validating

from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestRegressor


X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=42)


model = RandomForestRegressor(n_estimators=100, random_state=42)
cv_scores = cross_val_score(model,x,y,cv=5)

cv_mse_scores = cv_scores
mean_mse = cv_mse_scores.mean()

cv_scores_r2 = cross_val_score(model, x, y, cv=5, scoring='r2')
mean_r2 = cv_scores_r2.mean()

print("Mean Cross-Validation Mean Squared Error (MSE):", mean_mse)
print("Mean Cross-Validation R^2 Score:", mean_r2)
```
